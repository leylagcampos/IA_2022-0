# Detector de bordes de Canny

### Importar librerías

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

### Función para mostrar la imagen

```python
def mostrar_imagen(titulo, imagen):
    imagen = cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB)
    plt.imshow(imagen)
    plt.title(titulo)
    plt.grid(False)
    plt.show()
```

### Diccionario para los argumentos

```python
args = {"imagen":"lapiceros.jpg"}
```

### Carga la imagen y la muestra

```python
imagen = cv2.imread(args["imagen"])
mostrar_imagen("Original",imagen)
```

### Detección de bordes de Canny

Carga la imagen, la conviérte a escala de grises y la difumína ligeramente para eliminar los bordes de alta frecuencia que no nos interesan

```python
imagen = cv2.imread(args["imagen"])
imagen = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
imagen = cv2.GaussianBlur(imagen, (5, 5), 0)
mostrar_imagen("Difuminado", imagen)
```

Al realizar la detección de bordes de Canny, necesitamos dos valores para la histéresis: umbral1 y umbral2. Cualquier valor de gradiente mayor
que el umbral2 se considera un borde. Cualquier valor por debajo del umbral1 no se considera no borde. Los valores entre el umbral1 y el umbral2
se clasifican como borde o no borde en función de cómo están "conectadas" las intensidades. En este caso, cualquier valor de gradiente por debajo
de 30 se considera no borde, mientras que cualquier valor por encima de 150 se considera borde.

```python
canny = cv2.Canny(imagen, 30, 150)
mostrar_imagen("Canny", canny)
```
