- x= Inversión en publicidad en radio
- y= Total de ventas

### MODELO DE REGRESION LINEAL Simple

```python
x=datos[['Radio']]
```

```python
y= datos['Sales']
```

#### Regresión lineal con el Método de Mínimos Cuadrados

```python
from sklearn.linear_model import LinearRegression
```

```python
modelo = LinearRegression()
```

```python
modelo.fit(x,y)
```

##### Parámetros  obtenidos en el modelo

Interceptor Beta 0

```python
print(modelo.intercept_)
```

Otros Beta

```python
print(modelo.coef_)
```

#### Predicciones

```python
predicciones=modelo.predict(x)
```

```python
plt.scatter(y,predicciones)
```

```python
sns.regplot(x=y,y=predicciones,data=datos)
```

##### Métricas de evaluación

```python
from sklearn import metrics
MSE = metrics.mean_squared_error(y,predicciones)
RSS = MSE*200
print(RSS)
```
