# Deteccion de rostros

## Importar librerías

```python
from detector.detector_de_rostros import DetectorRostro 
import matplotlib.pyplot as plt
import cv2
```

## Función para mostrar imagen

```python
def mostrar_imagen(titulo, imagen):
 imagen = cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB)
 plt.imshow(imagen)
 plt.title(titulo)
 plt.grid(False)
 plt.show()
```

## Diccionario de argumentos

```python
args = {"imagen1":"imagenes/foto.jpg","imagen2":"imagenes/foto2.png",
            "rostro":"cascadas/haarcascade_frontalface_default.xml"}
```

## Carga la primera imagen y la muestra

```python
imagen1 = cv2.imread(args["imagen1"])
mostrar_imagen("Original",imagen1)
```

### Convertir la imagen a escala de grises

```python
gris = cv2.cvtColor(imagen1,cv2.COLOR_BGR2GRAY)
mostrar_imagen("Escala de grises",gris)
```

### Encontrar los rostros en la imagen

```python
dr = DetectorRostro(args["rostro"])
rectRostros = dr.detectar(gris,factorEscala=1.1, vecMin=5, tamMin=(30, 30))
print("Se encontró {} rostro(s)".format(len(rectRostros)))
```

### Dibujar los rectángulos sobre los rostros

```python
for(x , y, w, h) in rectRostros:
    cv2.rectangle(imagen1,(x,y),(x+w,y+h),(0,255,0),2)
```

```python
mostrar_imagen("Rostros",imagen1)
```
