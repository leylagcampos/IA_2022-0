# Suavizado y Difuminado

###  Importar las librerías

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

### Función para mostrar imágenes

```python
def mostrar_imagen(titulo, imagen):
    imagen = cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB)
    plt.imshow(imagen)
    plt.title(titulo)
    plt.grid(False)
    plt.show()
```

### Diccionario para establecer los argumentos

```python
args = {"imagen": "stich.png"}
```

### Cargar la imagen y mostrarla

```python
imagen = cv2.imread(args["imagen"])
mostrar_imagen("Original",imagen)
#cv2.imshow("Original",imagen)
#cv2.waitKey(0)
```

### Difuminado Promedio

```python
dif_prom01 = cv2.blur(imagen,(3,3))
mostrar_imagen("dif_prom01",dif_prom01)
#cv2.imshow("dif_prom01",dif_prom01)
#cv2.waitKey(0)
```

```python
dif_prom02 = cv2.blur(imagen,(5,5))
mostrar_imagen("dif_prom02",dif_prom02)
#cv2.imshow("dif_prom02",dif_prom02)
#cv2.waitKey(0)
```

```python
dif_prom03 = cv2.blur(imagen,(7,7))
mostrar_imagen("dif_prom03",dif_prom03)
```

### Difuminado Gausiano

```python
dif_gau01 = cv2.GaussianBlur(imagen,(3,3),0)
mostrar_imagen("dif_gau01",dif_gau01)
```

```python
dif_gau02 = cv2.GaussianBlur(imagen,(5,5),0)
mostrar_imagen("dif_gau02",dif_gau02)
```

```python
dif_gau03 = cv2.GaussianBlur(imagen,(7,7),0)
mostrar_imagen("dif_gau03",dif_gau03)
```

### Difuminado Mediano

```python
dif_med01 = cv2.medianBlur(imagen,3)
mostrar_imagen("dif_med01",dif_med01)
```

```python
dif_med01 = cv2.medianBlur(imagen,5)
mostrar_imagen("dif_med01",dif_med01)
```

```python
dif_med03 = cv2.medianBlur(imagen,7)
mostrar_imagen("dif_med03",dif_med03)
```

### Filtrado Bilateral

```python
fil01 = cv2.bilateralFilter(imagen,5,21,21)
mostrar_imagen("fil01",fil01)
```

```python
fil02 = cv2.bilateralFilter(imagen,7,31,31)
mostrar_imagen("fil02",fil02)
```

```python
fil03 = cv2.bilateralFilter(imagen,9,41,41)
mostrar_imagen("fil03",fil03)
```
