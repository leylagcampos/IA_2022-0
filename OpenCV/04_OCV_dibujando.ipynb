## Dibujando con OpenCV

### Importar las librerias

```python
# Importar las librerias necesarias
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

### Función para mostrar imágenes en Jupyter Notebooks y Google Colab

```python
def mostrar_imagen(titulo,imagen):
  imagen = cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB)
  plt.imshow(imagen)
  plt.title(titulo)
  plt.grid(False)
  plt.show
```

### Implementando funciones de dibujo básicas en OpenCV

```python
# inicializar el lienzo con una imagen de 300x300 pixeles con 3
# canales (Rojo, Verde y Azúl) con un fondo negro
canvas = np.zeros((300, 300, 3), dtype="uint8")
```

```python
# dibujar una línea verde desde la esquina superior izquierda del

# lienzo hasta la esquina inferior derecha
green = (0, 255, 0)
cv2.line(canvas, (0, 0), (300, 300), green)
mostrar_imagen("Canvas", canvas)
# dibujar una línea roja con 3 pixeles de grosor desde la esquina
# superior derecha a la esquina inferior izquierda
red = (0, 0, 255)
cv2.line(canvas, (300, 0), (0, 300), red, 3)
mostrar_imagen("Canvas", canvas)
```

```python
# dibujar un cuadrado verde de 50x50 pixeles, comenzando en 10,10 y

# terminando en 60,60
cv2.rectangle(canvas, (10, 10), (60, 60), green)
mostrar_imagen("Canvas", canvas)
# dibujar otro rectángulo, uno rojo con 5 pixeles de grosor
cv2.rectangle(canvas, (50, 200), (200, 225), red, 5)
mostrar_imagen("Canvas", canvas)
# dibujar un rectángulo y rellenarlo (azúl y rellenado )
blue = (255, 0, 0)
cv2.rectangle(canvas, (200, 50), (225, 125), blue, -1)
mostrar_imagen("Canvas", canvas)
```

```python
# reinicializar el canvas como un arreglo vacío, calcular el

# centro (x,y) del canvas
canvas = np.zeros((300, 300, 3), dtype="uint8")
(centerX, centerY) = (canvas.shape[1] // 2, canvas.shape[0] // 2)
white = (255, 255, 255)
# bucle para incrementar el radio, desde 25 pixeles hasta 150 pixeles
# en incrementos de 25
for r in range(0, 175, 25):
# draw a white circle with the current radius size
  cv2.circle(canvas, (centerX, centerY), r, white)
# mostrar el dibujo
mostrar_imagen("Canvas", canvas)
```

```python
# reinicializar el canvas nuevamente
canvas = np.zeros((300, 300, 3), dtype="uint8")
# dibujemos 25 circulos
for i in range(0, 25):
# generar un radio aleatorio entre 5 y 200, generar un color
# aleatorio, y luego escoger aleatoriamente un punto en el
# lienzo donde sera dibujado el círculo
  radius = np.random.randint(5, high=200)
  color = np.random.randint(0, high=256, size=(3,)).tolist()
  pt = np.random.randint(0, high=300, size=(2,))
  # dibujar el circulo aleatorio
  cv2.circle(canvas, tuple(pt), radius, color, -1)
# mostrar el dibujo
mostrar_imagen("Canvas", canvas)
```

### Dibujando en las imágenes con OpenCV

```python
args = {
"image": "lilostich.jpg"
}
```

```python
# cargar la imagen
image = cv2.imread(args["image"])
# dibujar un circulo, dos circulos rellenos y un rectángulo
cv2.circle(image, (168, 188), 90, (0, 0, 255), 2)
cv2.circle(image, (150, 164), 10, (0, 0, 255), -1)
cv2.circle(image, (192, 174), 10, (0, 0, 255), -1)
cv2.rectangle(image, (134, 200), (186, 218), (0, 0, 255), -1)
# mostrar la imagen
mostrar_imagen("Output", image)
```
