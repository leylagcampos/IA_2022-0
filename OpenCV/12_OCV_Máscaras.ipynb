# M치scaras con OpenCV

### Importar paquetes y librerias

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

### Funci칩n para mostrar una imagen con imshow() de pyplot

```python
def mostrar_imagen(title, imagen):
    imagen = cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB)
    plt.imshow(imagen)
    plt.title(title)
    plt.grid(False)
    plt.show()
```

## M치scaras

```python
args = { "imagen" :"reptil.jpg" }
```

```python
# cargar la imagen original y mostrarla
imagen = cv2.imread(args["imagen"])
mostrar_imagen("Original",imagen)
```

La m치scara tiene el mismo tama침o de la imagen, pero solamente tiene dos valores de pixel, 0 y 255. Los pixeles de la m치scara con valor 0 son
ignorados en la imagen original, mientras que los pixeles de la m치scara con valores 255 se conservan.

```python
imagen.shape[:2]
```

#### M치scara Rectangular

```python
mascara = np.zeros(imagen.shape[:2],dtype="uint8")
cv2.rectangle(mascara,(0,0),(150,150),255,-1)
mostrar_imagen("M치scara Rectangular", mascara)
```

```python
conMascara = cv2.bitwise_and(imagen,imagen,mask=mascara)
mostrar_imagen("M치scara aplicada a la imagen",conMascara)
```

#### M치scara Circular

Esta m치scara circular tiene un radio igual a 50 pixeles

```python
mascara = np.zeros(imagen.shape[:2],dtype="uint8")
cv2.circle(mascara,(50,50),50,255,-1)
mostrar_imagen("M치scara Circular", mascara)
```

```python
conMascara = cv2.bitwise_and(imagen,imagen,mask=mascara)
mostrar_imagen("M치scara aplicada a la imagen",conMascara)
```
